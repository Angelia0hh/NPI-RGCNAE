```python
import os
import sys
import pandas as pd
from pandas import DataFrame
import csv
```

```python
path = 'D:\\PycharmProject\\GCNAE\\result\\'
# 将数据前的tensor去掉
for data in ['NPInter_10412','RPI2241', 'RPI369', 'RPI7317']:
    pathdir = os.listdir(path+data+'\\')
    print(data)
    
    for f in pathdir:
        if f.startswith('metrics'):
            result = []
            with open(path+data+'\\'+f,'r+')as file:
                for line in file:
                    line = line.rstrip().split(',')           
                    if line[0].startswith('fold'):
                        line[1:-1] = [num[7:-1]for num in line[1:-1]]
                        result.append(line)
                    else:
                        result.append(line)
            #print(result)
            with open(path+data+'\\'+f, "w",newline='') as file:
                writer = csv.writer(file)
                for i in result:
                    writer.writerow(i)
```

```python
#计算不同数据集上10次5折的平均值
for data in ['NPInter_10412','RPI2241', 'RPI369', 'RPI7317']:
    pathdir = os.listdir(path+data+'\\')
    print(data)
    for f in pathdir:
        if f.startswith('metrics'):
            if len(f.split('_'))==6:
                method = 'sort_random'
            else:
                method = f.split('_')[1]
            print(method)
            dataframe = pd.read_csv(path+data+'\\'+f)
            dataframe.columns = ['name','acc','pre','sen','spe','MCC','FPR','TPR','AUC']
            all_avg = dataframe[(dataframe['name']=='average')]
            all_avg = all_avg[['acc','pre','sen','spe','MCC','FPR','TPR','AUC']].astype(float)
            std = list(all_avg[['acc','pre','sen','spe','MCC','FPR','TPR','AUC']].std())
            std = [ format(i, '.7f') for i in std]
            std.insert(0,'std')
            mean = list(all_avg[['acc','pre','sen','spe','MCC','FPR','TPR','AUC']].mean())
            mean.insert(0,'mean')
            #在原来的文件中追加mean和std
            with open(path+data+'\\'+f,'a',newline='')as file:
                csv_write = csv.writer(file,dialect='excel')
                csv_write.writerow(mean)#每个数据集上十次五折的平均结果(stack+node2vec+side)
                csv_write.writerow(std)#十次五折的方差
            #将不同数据集上的mean和std汇总到一个表中
            with open('D:\\PycharmProject\\GCNAE\\comparison\\5-fold of LPI-RGCNAE on different datasets.csv','a',newline='')as f:
                # stack+node2vec+side在不同数据集和不同负样本生成方式上的十次五折平均值及方差
                csv_write = csv.writer(f,dialect='excel')
                mean.insert(0,method)
                mean.insert(0,data)
                csv_write.writerow(mean)#每个数据集上十次五折的平均结果
                std.insert(0,method)
                std.insert(0,data)
                csv_write.writerow(std)#每个数据集上十次五折的平均结果
```

```python
case_study_RNA = ['NEAT1','MALAT1']

RPI7317 = pd.read_csv('data/raw_data/RPI7317.csv')
protein_list = RPI7317['Protein names'].unique().tolist()
ncRNA_list = RPI7317['RNA names'].unique().tolist()

RNA_dicts = {}
for RNA, protein in RPI7317['Protein names'].groupby(RPI7317['RNA names']):
    RNA_dicts[RNA] = list(protein)
```

```python
edgelist = pd.read_csv('D://PycharmProject//GCNAE//data//generated_data//RPI7317//edgelist_sort.csv',header=None)
NPI_pos_matrix = pd.read_csv('data//generated_data//RPI7317//NPI_pos.csv', header=None).values
NPI_neg_matrix = pd.read_csv("data//generated_data//RPI7317//NPI_neg_sort.csv", header=None).values
```

```python
for case in case_study_RNA:
    # 找到所有已知和cast_study 关联的对
    pos_related_pairs = [] # 和case RNA所有相关的正样本 index对
    for pr in RNA_dicts[case]:
        pos_related_pairs.append([ncRNA_list.index(case), protein_list.index(pr)])
    
    print(len(pos_related_pairs))

    # 找到所有需要在训练集中去掉的RNA-蛋白质对的下标
    delete_pairs = pos_related_pairs[len(pos_related_pairs)//3:] # 去掉1/3的相关的正样本
    delete_pairs = np.array(delete_pairs)
    
    # 将case_study中对的信息从训练集中去掉
    mask = np.ones((len(ncRNA_list), len(protein_list)))
    mask[delete_pairs[:,0],delete_pairs[:,1]] = 0  # 将验证集中的边置0
    print("raw dataset:")
    print(NPI_pos_matrix.sum())
    print(NPI_neg_matrix.sum())
    tmp_pos = NPI_pos_matrix * mask
    tmp_neg = NPI_neg_matrix * mask
    
    print("The number of negative samples in the matrix:")
    print(tmp_neg.sum())
    print("The number of positive samples in the matrix:")
    print(tmp_pos.sum())
    
```
