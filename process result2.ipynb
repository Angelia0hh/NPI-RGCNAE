```python
import os
import sys
import pandas as pd
from pandas import DataFrame
import csv
```

```python
path = 'D:\\PycharmProject\\GCNAE\\results_3.11\\'
# 将数据前的tensor去掉
for data in ['NPInter_10412','RPI2241', 'RPI369', 'RPI7317']:
    pathdir = os.listdir(path+data+'\\negative_method\\')
    print(data)
    
    for f in pathdir:
        if f.startswith('metrics'):
            result = []
            with open(path+data+'\\negative_method\\'+f,'r+')as file:
                print(path+data+f)
                for line in file:
                    line = line.rstrip().split(',')           
                    new_line = [item[7:-1] for item in line[:-1]]
                    new_line.append(line[-1])
                    result.append(new_line)
            print(result)
            with open(path+data+'\\negative_method\\'+f, "w",newline='') as file:
                writer = csv.writer(file)
                for i in result:
                    writer.writerow(i)    
```

```python
#计算不同数据集上10次5折的平均值
pd.set_option('precision', 6)
for data in ['NPInter_10412','RPI2241', 'RPI369', 'RPI7317']:
    pathdir = os.listdir(path+data+'\\negative_method\\')
    print(data)
    for f in pathdir:
        if f.startswith('metrics'):
            method = f.split('_')[1]
            print(method)
            dataframe = pd.read_csv(path+data+'\\negative_method\\'+f,header=None)   
            std = dataframe.std()
            std = [ format(i, '.7f') for i in std]
            std.insert(0,'std')
            mean = list(dataframe.mean())
            mean.insert(0,'mean')
            print(mean)
            print(std)
            
            #将不同数据集上的mean和std汇总到一个表中
            with open('D:\\PycharmProject\\GCNAE\\tables_3.11\\negative sample generation method comparison of RGCNAE.csv','a',newline='')as f:
                # sort_stack_random_side在不同数据集和不同负样本生成方式上的十次五折平均值及方差
                csv_write = csv.writer(f,dialect='excel')
                mean.insert(0,method)
                mean.insert(0,data)
                csv_write.writerow(mean)#每个数据集上十次五折的平均结果
                std.insert(0,method)
                std.insert(0,data)
                csv_write.writerow(std)#每个数据集上十次五折的平均结果
```

```python
#不同的特征组合结果
for data in ['NPInter_10412', 'RPI369']:
    i=0
    pathdir = os.listdir(path+data+'\\combination\\')
    # print(pathdir)
    for file in pathdir:
        if file.startswith('metrics'):
            method = file.split('_')
            combination = method[2]+"+"+method[4]
            print(combination)
            result = []
        
            with open(path+data+'\\combination\\'+file,'r+')as f:
                for line in f:
                    line = line.rstrip().split(',')           
                    new_line = [item[7:-1] for item in line[:-1]]
                    new_line.append(line[-1])
                    result.append(new_line)
            #print(result)
            
            with open(path+data+'\\combination\\'+file, "w",newline='') as f:
                writer = csv.writer(f)
                for i in result:
                    writer.writerow(i)
            
            
            dataframe = pd.read_csv(path+data+'\\combination\\'+file,header=None)
            std = list(dataframe.std())
            std = [ format(i, '.7f') for i in std]
            std.insert(0,'std')
            mean = list(dataframe.mean())
            mean.insert(0,'mean')
            print(std)
            print(mean)
            
        
            #将不同数据集上的mean和std汇总到一个表中
            with open('tables_3.11\\different combinations comparison on '+data+'.csv','a',newline='')as f:
                csv_write = csv.writer(f,dialect='excel')
                if i==0:
                    csv_write.writerow([data])
                    i+=1
                mean.insert(0,combination)
                csv_write.writerow(mean)#每个数据集上十次五折的平均结果
                std.insert(0,combination)
                csv_write.writerow(std)#每个数据集上十次五折的平均结果
```
